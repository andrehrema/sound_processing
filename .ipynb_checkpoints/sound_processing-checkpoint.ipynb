```python
from pydub import AudioSegment
from matplotlib import pyplot as plt
import numpy as np
from pydub.playback import play
import audiosegment
import sounddevice as sd
```

```python
sound = AudioSegment.from_mp3("snare.mp3")
raw_data = sound.get_array_of_samples()

size_raw_data = len(raw_data)
print("raw_data size: "+ str(size_raw_data))

sample_rate = sound.frame_rate
print("The sample rate is: "+str(sample_rate))

print("Durantion of sound: "+str(size_raw_data/sample_rate))
```

```python
t = np.linspace(0, size_raw_data, size_raw_data)
exp_factor = np.exp(-5*(t/sample_rate))
new_sound = raw_data * exp_factor

exp_factor[0:10]
```

```python
plt.subplot(121)
plt.plot(raw_data)
plt.subplot(122)
plt.plot(new_sound)
plt.show()
```

```python
upper_limit = round(size_raw_data*0.5)
fourier_raw_data = np.fft.fft(raw_data, upper_limit)
magnitude_fourier_db = 20*np.log(np.abs(fourier_raw_data))
freq = np.fft.fftfreq(upper_limit)
plt.plot(freq, magnitude_fourier_db)
plt.show()
```

```python
fourier_new_sound = np.fft.fft(new_sound, upper_limit)
new_magnitude_fourier_db = 20*np.log(np.abs(fourier_new_sound))
freq = np.fft.fftfreq(upper_limit)
plt.plot(freq, new_magnitude_fourier_db)
plt.show()
```

```python
new_sound = np.array(new_sound, dtype=np.int32)
new_sound[:5]
```

```python
new_sound_raw = audiosegment.from_numpy_array(new_sound, sample_rate)
play(new_sound_raw)
```
